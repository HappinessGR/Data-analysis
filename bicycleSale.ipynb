# Visualization

```python
df.plot()
```

```python
grouped = df.groupby('Country')['Unit_Price'].sum()

# Horizontal bar chart
grouped.plot.barh(title='Total Unit_Price by Region', color='darkblue')
plt.xlabel('Sales')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
sales_by_country = df.groupby('Country')['Unit_Price'].sum()

# Create a pie chart
sales_by_country.plot.pie(
    figsize=(10, 7),
    autopct='%1.1f%%',
    startangle=90,
    ylabel='',  # removes y-axis label
    title='Sales Distribution by Country'
)
plt.show()
```
